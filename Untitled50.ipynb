```python
from  scipy import stats
```

```python
df= pd.read_csv(r'C:\Users\Naren karthikeyan\Downloads\archive (2)\dataco_clean.csv')
df.head()
```

```python
print("TEST 1 — ANOVA: Does delay_days vary significantly by region?\n")
regional_groups = [
    group['delay_days'].dropna().values
    for name, group in df.groupby('order_region')
]
f_stat, p_value = stats.f_oneway(*regional_groups)
print(f"F-statistic: {f_stat:.4f}")
print(f"P-value:     {p_value:.6f}")
print(f"Significant: {'YES' if p_value < 0.05 else 'NO'} (α = 0.05)")

print("\nTEST 2 — CHI-SQUARE: Is shipping mode associated with delay?")
contingency = pd.crosstab(df['shipping_mode'], df['is_delayed'])
print("Contingency Table:")
print(contingency)
chi2, p_chi, dof, expected = stats.chi2_contingency(contingency)
print(f"\nChi-square statistic: {chi2:.4f}")
print(f"Degrees of freedom:   {dof}")
print(f"P-value:              {p_chi:.6f}")
print(f"Significant:          {'YES' if p_chi < 0.05 else 'NO'} (α = 0.05)")

print("\nTEST 3 — KRUSKAL-WALLIS: Does delay vary by product category?")
category_groups = [
    group['delay_days'].dropna().values
    for name, group in df.groupby('category_name')
]
h_stat, p_kruskal = stats.kruskal(*category_groups)
print(f"H-statistic: {h_stat:.4f}")
print(f"P-value:     {p_kruskal:.6f}")
print(f"Significant: {'YES' if p_kruskal < 0.05 else 'NO'} (α = 0.05)")

print("\nTEST 4 — Z-TEST: Is Standard Class delay rate significantly")
print("         lower than Same Day?")

standard = df[df['shipping_mode'] == 'Standard Class']['is_delayed']
sameday  = df[df['shipping_mode'] == 'Same Day']['is_delayed']

n1, p1 = len(standard), standard.mean()
n2, p2 = len(sameday),  sameday.mean()
p_pool = (standard.sum() + sameday.sum()) / (n1 + n2)
z_stat = (p1 - p2) / np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))
p_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Standard Class delay rate: {p1:.4f} ({p1*100:.2f}%)")
print(f"Same Day delay rate:       {p2:.4f} ({p2*100:.2f}%)")
print(f"Z-statistic:               {z_stat:.4f}")
print(f"P-value:                   {p_z:.6f}")
print(f"Significant:               {'YES' if p_z < 0.05 else 'NO'} (α = 0.05)")

print("\nTEST 5 — CORRELATION: Does order value correlate with delay cost?")
corr_df = df[['order_item_total', 'delay_cost']].dropna()
pearson_r,  p_pearson  = stats.pearsonr(corr_df['order_item_total'], 
                                         corr_df['delay_cost'])
spearman_r, p_spearman = stats.spearmanr(corr_df['order_item_total'], 
                                          corr_df['delay_cost'])
print(f"Pearson  r:  {pearson_r:.4f}  | p-value: {p_pearson:.6f}")
print(f"Spearman r:  {spearman_r:.4f}  | p-value: {p_spearman:.6f}")
print(f"Significant: {'YES' if p_pearson < 0.05 else 'NO'} (α = 0.05)")
```
